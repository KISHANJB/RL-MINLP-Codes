```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV (row 1 will be used as header automatically)
df = pd.read_csv("/home/kkishan/Desktop/QFunction Approximator CSVs/results.csv")

# Extract column names
x_col = df.columns[0]   # first column name
y_col = df.columns[1]   # second column name

block_size = 100
n = len(df)

for i in range(0, n, block_size):
    block = df.iloc[i:i+block_size]
    plt.plot(block[x_col], block[y_col], label=f'Block {i//block_size + 1}')

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title("Curves for Batch.nl (100-row segments)")
plt.legend()
plt.show()
```
